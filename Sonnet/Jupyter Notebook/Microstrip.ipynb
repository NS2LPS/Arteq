```python
import matplotlib.pyplot as plt
import numpy as np
import skrf as rf
from scipy.optimize import curve_fit
```

## Simuler une microstrip pour 3 largeurs différentes 

- Utiliser une boîte de $10\,\mathrm{mm} \times 2\,\mathrm{mm}$ avec une *cell size* adaptée à la géométrie de la ligne,
- Charger les 3 fichiers S2P générés par Sonnet.


```python
path="..//Sonnet"  # A changer: indiquer le répertoire dans lequel sont sauvé les fichier .snp générés par Sonnet

MicroStrip = []
for i in range(1,4):
    file_name =path+"/Microstrip_param"+ str(i)+".s2p"
    MicroStrip.append(rf.Network(file_name))
```

```python
Param=['100 $\\mu m$','200 $\\mu m$','300 $\\mu m$']
```

```python
plt.figure(1)
for i in range(0,3):
    MicroStrip[i].plot_s_db(0,0, label=Param[i])
    
```
